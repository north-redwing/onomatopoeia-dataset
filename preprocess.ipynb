```
import os
from pydub import AudioSegment
from pydub.silence import split_on_silence
from pydub.playback import play
```

```
dir_names = ['train', 'test', 'valid']
for dir_name in dir_names:
    filenames = os.listdir(dir_name)
    for filename in filenames:
        data = AudioSegment.from_file(os.path.join(dir_name, filename), format='wav')
        chunks = split_on_silence(data, min_silence_len=100, silence_thresh=-60, keep_silence=100)
        for i, chunk in enumerate(chunks):
            if i == 0: chunk.export(os.path.join(dir_name, filename), format='wav')
            else: chunk.export(os.path.join(dir_name, str(i)+'_'+filename), format='wav')
```
